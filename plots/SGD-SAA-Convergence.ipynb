```python
# True weights and starting weights

np.random.seed(17)

d1 = 10
d2 = 1

m_true = 20
A_true = np.random.randn(m_true,d1)*np.sqrt(2/m_true)
B_true = np.random.randn(m_true,m_true)*np.sqrt(2/m_true)
C_true = np.random.randn(d2,m_true)*np.sqrt(1/d2)

m = 10
A0 = np.random.randn(m,d1)*np.sqrt(2/m)
B0 = np.random.randn(m,m)*np.sqrt(2/m)
C0 = np.random.randn(d2,m)*np.sqrt(1/d2)

def relu(x):
    y = np.maximum(0,x)
    return y

n = 250
X = np.random.randn(n,d1)
X /= la.norm(X,axis=1).reshape(n,1)
Y = np.transpose(C_true@relu(B_true@relu(A_true@X.T)))
```

```python
def grad(A,B,C,x,y):
    m = np.shape(B)[0]
    D1 = np.zeros(m)
    D2 = np.zeros(m)
    z = A@x
    D1[z >= 0] = 1
    D1 = np.diag(D1)
    D1Ax = D1@z
    z = B@D1Ax
    D2[z >= 0] = 1
    D2 = np.diag(D2)
    D2BD1Ax = D2@z
    z = C@D2BD1Ax-y
    obj = .5*la.norm(z)**2
    CD2 = C@D2
    CD2BD1 = CD2@B@D1
    der_A = np.outer(CD2BD1.T@z,x)
    der_B = np.outer(CD2.T@z,D1Ax)
    der_C = np.outer(z,D2BD1Ax)
    return [obj,der_A,der_B,der_C]

def gradset(A,B,C,X,Y):
    # TODO: can we vectorize this??
    
    n = np.shape(X)[0]
    objset = 0
    derset_A = np.zeros(np.shape(A))
    derset_B = np.zeros(np.shape(B))
    derset_C = np.zeros(np.shape(C))
    for i in range(n):
        x = X[i,:]
        y = Y[i,:]
        [obj,der_A,der_B,der_C] = grad(A,B,C,x,y)
        objset += obj
        derset_A += der_A
        derset_B += der_B
        derset_C += der_C
    objset /= n
    derset_A /= n
    derset_B /= n
    derset_C /= n
    return [objset,derset_A,derset_B,derset_C]
```

```python
def sgd(A, B, C, X, Y, step_size, niter, batch_size, n_test, niter_test):
    
    
    # draw niter_test iterates
    p = np.arange(niter)
    p = 1/np.sqrt(p+1)
    p = p/np.sum(p)
    sampled_iterates = np.random.choice(niter,size=niter_test,replace=True,p=p)
    sampled_iterates = np.unique(sampled_iterates)
    sampled_iterate_index = 0
    sampled_iterate = sampled_iterates[sampled_iterate_index]
    sampled_iterate_final = sampled_iterates[len(sampled_iterates)-1]
    
    
    # draw test and validation sets
    choice = np.random.choice(n,size=n_test,replace=True)
    X_test = X[choice,:]
    Y_test = Y[choice]
    

    # apply sgd
    error = np.inf
    t_err = np.inf
    for i in range(niter):
        
        # draw stochastic gradient
        choice = np.random.choice(n,size=batch_size,replace=True)
        X_batch = X[choice,:]
        Y_batch = Y[choice]
        if batch_size == 1:
            [obj,der_A,der_B,der_C] = grad(A,B,C,X_batch,Y_batch)
        else:
            [obj,der_A,der_B,der_C] = gradset(A,B,C,X_batch,Y_batch)
        
        # take gradient step
        eta = step_size/np.sqrt(i+1)
        A = A-eta*der_A
        B = B-eta*der_B
        C = C-eta*der_C
        
        # compute test error for sampled iterates
        if i == sampled_iterate:
            [obj,der_A,der_B,der_C] = gradset(A,B,C,X_test,Y_test)
            new_error = la.norm(der_A)**2+la.norm(der_B)**2+la.norm(der_C)**2
            if new_error < error:
                error = new_error
                A_best = A
                B_best = B
                C_best = C
            if sampled_iterate != sampled_iterate_final:
                sampled_iterate_index += 1
                sampled_iterate = sampled_iterates[sampled_iterate_index]
        
        # compute the true error of each iterate
        [obj,der_A,der_B,der_C] = gradset(A,B,C,X,Y)
        new_error = la.norm(der_A)**2+la.norm(der_B)**2+la.norm(der_C)**2
        if new_error < t_err:
            t_err = new_error
        
        
    # compute validation error for best iterate and last iterate
    [obj,der_A,der_B,der_C] = gradset(A_best,B_best,C_best,X,Y)
    b_err = la.norm(der_A)**2+la.norm(der_B)**2+la.norm(der_C)**2
    [obj,der_A,der_B,der_C] = gradset(A,B,C,X,Y)
    l_err = la.norm(der_A)**2+la.norm(der_B)**2+la.norm(der_C)**2
    
    return [b_err,t_err,l_err]
```

```python
step_size = .1
niter = 1000
batch_size = 10
n_test = 100
niter_test = 100

num_trials = 100
b_errs = np.zeros(num_trials)
t_errs = np.zeros(num_trials)
l_errs = np.zeros(num_trials)
for i in range(num_trials):
    print(i)
    A = A0
    B = B0
    C = C0
    [b_err,t_err,l_err] = sgd(A, B, C, X, Y, step_size, niter, batch_size, n_test, niter_test)
    b_errs[i] = b_err
    t_errs[i] = t_err
    l_errs[i] = l_err

np.savetxt('saa_conv_b_errs.csv', b_errs, delimiter=',')
np.savetxt('saa_conv_t_errs.csv', t_errs, delimiter=',')
np.savetxt('saa_conv_l_errs.csv', l_errs, delimiter=',')
```

```python
# b_errs = np.loadtxt('saa_conv_b_errs.csv',  delimiter=',')
# t_errs = np.loadtxt('saa_conv_t_errs.csv',  delimiter=',')
# l_errs = np.loadtxt('saa_conv_l_errs.csv',  delimiter=',')
b_errs = np.loadtxt('data/saa_conv_b_errs.csv',  delimiter=',')
t_errs = np.loadtxt('data/saa_conv_t_errs.csv',  delimiter=',')
l_errs = np.loadtxt('data/saa_conv_l_errs.csv',  delimiter=',')
cdf = np.arange(len(b_errs))/float(len(b_errs))
b_errs = np.sort(b_errs)
t_errs = np.sort(t_errs)
l_errs = np.sort(l_errs)
```

```python
# Find line of best-fit (possibly after doing a transformation)
def bestFit( x, y):
    """ returns slope,intercept for best fit of the data (x,y)"""
    # Form Vandermonde matrix
    n = len(x)
    V = np.vstack( (x.ravel(), np.ones(n)) ).T 
    
    beta = la.lstsq(V,y,rcond=None)[0]

    slope = beta[0]
    intercept = beta[1]

    return slope,intercept
```

```python
# Figure

q = 30
deltas = 1-cdf[-q:] # behavior of delta near 1 is not interesting; we care about delta --> 0
data_b = b_errs[-q:]
data_t = t_errs[-q:]
data_l = l_errs[-q:]


fig, ax = plt.subplots()
ax.semilogx(deltas,data_b,'-',label='post-processing iterate')
ax.semilogx(deltas,data_t,'--',label='best iterate')
ax.semilogx(deltas,data_l,'-.',label='last iterate')


plt.xlabel('Failure probability $\delta$')
plt.ylabel('($1-\delta$)-percentile error')
plt.ylim(bottom=0)
legend = ax.legend(loc='lower right')
# plt.savefig('saaconvergence_v2.pdf')
plt.show()
```

```python
q = 20
deltas = 1-cdf[-q:] # behavior of delta near 1 is not interesting; we care about delta --> 0
data_b = b_errs[-q:]
data_t = t_errs[-q:]
data_l = l_errs[-q:]


fig, ax = plt.subplots()#1,1,figsize=(4,6)) # default is (4,6)
fig.set_figwidth(9.6)
h1=ax.semilogx(deltas,data_b,'o-',markersize=5,label='post-processing iterate')
h2=ax.semilogx(deltas,data_t,'s-',markersize=5,label='best iterate')
h3=ax.semilogx(deltas,data_l,'d-',markersize=5,label='last iterate')

# Make a line of best fit, using function dependence f(delta)
f = lambda delta : np.log(1/delta)
clr1 = h1[0].get_color()
clr2 = h2[0].get_color()
clr3 = h3[0].get_color()
slp,intrcpt = bestFit( f(deltas), data_b)
ax.semilogx( deltas, slp*f(deltas) + intrcpt, '--', color=clr1,\
    markevery=.45,label="$\log(\delta^{-1})$ best fit, post-proc")
slp,intrcpt = bestFit( f(deltas), data_t)
ax.semilogx( deltas, slp*f(deltas) + intrcpt, '-.', color=clr2,\
    markevery=.45,label="$\log(\delta^{-1})$ best fit, best iter")
slp,intrcpt = bestFit( f(deltas), data_l)
ax.semilogx( deltas, slp*f(deltas) + intrcpt, ':', color=clr3,\
    markevery=.45,label="$\log(\delta^{-1})$ best fit, last iter")

plt.xlabel('Failure probability $\delta$')
plt.ylabel('($1-\delta$)-percentile error')
# plt.ylim(bottom=0)
# legend = ax.legend(loc='upper right')
plt.legend(bbox_to_anchor=(1.04,1), loc="upper left",borderaxespad=0)

# plt.savefig('saaconvergence_v2_zoomed.pdf')
plt.show()
```
